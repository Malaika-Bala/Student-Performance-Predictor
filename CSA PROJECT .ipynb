```python
print("=== Student Performance Predictor ===\n")

# Take input from user
study_hours = float(input("Enter study hours: "))
attendance = float(input("Enter attendance (%): "))
previous_marks = float(input("Enter previous marks: "))

# Simple formula (manually designed)
# You can explain this in viva easily
predicted_marks = (
    (study_hours * 5) +        # study impact
    (attendance * 0.3) +      # attendance impact
    (previous_marks * 0.5)    # past performance impact
)

# Limit marks to 100
if predicted_marks > 100:
    predicted_marks = 100

# Show result
print("\nPredicted Final Marks:", round(predicted_marks, 2))
```
